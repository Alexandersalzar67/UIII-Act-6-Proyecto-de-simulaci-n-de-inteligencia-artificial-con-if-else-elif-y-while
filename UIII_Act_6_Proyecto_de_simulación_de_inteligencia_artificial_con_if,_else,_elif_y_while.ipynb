```python
# ============================================
# SISTEMA DE IA - SCORE CREDITICIO
# Nivel: Principiante
# Uso de:
# - if
# - elif
# - else
# - while
# ============================================

print("===================================")
print(" IA DE ANALISIS DE SCORE CREDITICIO ")
print("===================================")

continuar = "si"

while continuar == "si":

    print("\nIngrese los datos del cliente\n")

    nombre = input("Nombre del cliente: ")
    edad = int(input("Edad: "))
    ingresos = float(input("Ingresos mensuales: "))
    deudas = float(input("Total de deudas: "))
    pagos = input("¿Paga a tiempo? (si/no): ")
    tarjetas = int(input("Cantidad de tarjetas de credito: "))

    score = 0

    # --------------------------------
    # ANALISIS DE EDAD
    # --------------------------------
    if edad >= 18:
        score = score + 20
        print("Edad valida para credito")
    else:
        score = score - 50
        print("No cumple la edad minima")

    # --------------------------------
    # ANALISIS DE INGRESOS
    # --------------------------------
    if ingresos >= 30000:
        score = score + 40
        print("Ingresos altos")
    elif ingresos >= 15000:
        score = score + 25
        print("Ingresos medios")
    else:
        score = score + 10
        print("Ingresos bajos")

    # --------------------------------
    # ANALISIS DE DEUDAS
    # --------------------------------
    if deudas <= 5000:
        score = score + 30
        print("Nivel de deuda bajo")
    elif deudas <= 20000:
        score = score + 15
        print("Nivel de deuda medio")
    else:
        score = score - 20
        print("Nivel de deuda alto")

    # --------------------------------
    # ANALISIS DE PAGOS
    # --------------------------------
    if pagos == "si":
        score = score + 35
        print("Buen historial de pagos")
    else:
        score = score - 30
        print("Mal historial de pagos")

    # --------------------------------
    # ANALISIS DE TARJETAS
    # --------------------------------
    if tarjetas <= 2:
        score = score + 20
        print("Buen control financiero")
    elif tarjetas <= 5:
        score = score + 10
        print("Uso moderado de tarjetas")
    else:
        score = score - 15
        print("Muchas tarjetas registradas")

    # --------------------------------
    # RESULTADO FINAL
    # --------------------------------
    print("\n==============================")
    print("RESULTADO DEL ANALISIS")
    print("==============================")

    print("Cliente:", nombre)
    print("Score obtenido:", score)

    if score >= 100:
        print("Clasificacion: EXCELENTE")
        print("Credito aprobado")
    elif score >= 70:
        print("Clasificacion: BUENO")
        print("Credito aprobado con revision")
    elif score >= 40:
        print("Clasificacion: REGULAR")
        print("Credito en observacion")
    else:
        print("Clasificacion: RIESGO ALTO")
        print("Credito rechazado")

    # --------------------------------
    # REPETIR PROGRAMA
    # --------------------------------
    continuar = input("\n¿Desea analizar otro cliente? (si/no): ")

print("\nPrograma finalizado.")
```
